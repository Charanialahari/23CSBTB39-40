```python
import pandas as pd
#load your excel data
file_path='salary.xls'
#update with your file path
data=pd.read_excel(file_path)
#display the first few rows of the data
print(data.head())
#Assuming the last column is the largest variable and others are features
x=data.iloc[:,:-1]#Features
y=data.iloc[:,-1]#Target
#Display the features and target
print("Features:\n",x.head())
print("Target:\n",y.head())
y = data.iloc[:, -1] #Target
import matplotlib.pyplot as plt
#Plot each feature against the target
for column in x.columns:
    plt.figure(figsize=(8,6))
    plt.scatter(x[column], y, alpha=0.5)
    plt.title(f'{column} vs Target') # Fixed f-string formatting
    plt.xlabel(column)
    plt.show() # Dedent this line to align with the 'for' loop
from sklearn.preprocessing import MinMaxScaler
#Create a scaler object
scaler = MinMaxScaler()
#Normalize the features and target
x_normalized = scaler.fit_transform(x) # Fix variable name and method call
y_normalized = scaler.fit_transform(y.values.reshape(-1, 1)) # Fix variable name, method call, and reshape
# Convert back to DataFrame for easier handling
x_normalized = pd.DataFrame(x_normalized, columns=x.columns) # Fix variable names and capitalization
y_normalized = pd.Series(y_normalized.flatten(), name='Target') # Fix variable names and method call
# Display the normalized data
print("Normalized Features \n", x_normalized.head()) # Fix variable name and method call
```
