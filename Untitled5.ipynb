```python
from queue import Queue

graph = {0: [1,2,3],
         1: [0,2],
         2: [1,3,5],
         3:[0,2,4],
         4:[3,5,7],
         5:[2,4,6,7],
         6:[5,7],
         7:[]}
print("The adjecency list representing the graph is:")
print(graph)

def bfs(graph, source):
    Q = Queue()
    Q.put(source)
    visited_vertices = {source} # Initialize visited_vertices as a set

    while not Q.empty(): # Use Q.empty() to check if the queue is empty
        vertex = Q.get()
        print(vertex,end="-->")
        for u in graph[vertex]:
            if u not in visited_vertices:
                Q.put(u)
                visited_vertices.add(u) # Use add method to update set

print("BFS traversal of graph with source 1 is:")
bfs(graph,1)
```
